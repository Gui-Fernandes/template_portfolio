 Series é um array(lista de valores) rotulado unidimensional capaz de conter dados de qualquer tipo (inteiro, string, float, objetos python, etc.). Os rótulos dos eixos são chamados coletivamente de índice.


```
valores = pd.Series([1, 6, 8, 12, 5])

valores
```

 A coluna da esquerda é chamada de ìndice, e o da direita são os próprios valores.

Para que os valores numéricos sejam mostrados em um array(lista), e mostrar o tamnho da coluna(stop) deve-se executar os seguintes códigos:

```
valores.values
```

```
valores.index
```

 É possível também atribuir nomes aos índices:

```
valores = pd.Series([5,10,15,10,5], index=["Will", "Charles", "Richard", "Ribamar", "Cleber"])
valores
```

O index nos ajuda para referenciar um determinado valor, ele nos permite acessar os valores pelo seu rótulo:

```
valores['Will']
```

 Outra ultilidade dessa biblioteca é gerar informações matemáticas dos dados, como média dos valores (.mean()) e desvio padrão.(.std()). 

```
print("Média:", valores.mean())
print("Desvio padrão:", valores.std())
```

Geralmente para gerar de forma breve as estatísticas, se usa o .describe()

```
valores.describe()
```

# Gerando um DataFrame

Um DataFrame é uma estrutura de dados bidimensional, ou seja, os dados são alinhados de forma tabular em linhas e colunas.

# Recursos do DataFrame

*   As colunas podem muitas vezes serem de tipos diferentes
*   Tamanho de linhas e colunas são mutáveis
*   Possui eixos rotulados (linhas e colunas)
*   Pode realizar operações aritméticas em linhas e colunas

# Criando um DataFrame a partir de uma lista

O DataFrame pode ser criado usando uma única lista ou uma lista de listas.



```
lista = [2, 4, 6, 8, 10]
df = pd.DataFrame(lista)
print(df)
```

Atribuir nome para as colunas:

```
lista = [['Mario',10],['João',12],['Mariana',13]]
df = pd.DataFrame(lista,columns=['Nome','Idade'])
print(df)
```

Para selecionar pelo index ou rótulo usamos o atributo .loc:

```
df.loc[2]
```
